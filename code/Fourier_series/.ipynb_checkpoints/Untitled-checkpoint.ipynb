```julia
using FFTW, Plots

function plot_sol(bc,bd, I=[0,2π])
    a = I[1]; b = I[2]
    N = (length(bc)+1)/2 # 2N-1
    n_pad = 200
    bc_pad = [zeros(n_pad);bc;zeros(n_pad)]
    bd_pad = [zeros(n_pad);bd;zeros(n_pad)]
    N_pad = N + n_pad
#     h_pad = (b-a)/(2N_pad-1)
#     xj_pad = a .+ h_pad*(0:2N_pad-2)
　　xNj_pad = real((2N_pad-1)*ifft(ifftshift(bc_pad)))
　　yNj_pad = real((2N_pad-1)*ifft(ifftshift(bd_pad)))
　　plot(xNj_pad, yNj_pad, legend=false,
　　    xlabel = "\$x\$", ylabel = "\$y\$")
end

N = 5
p = 5
m = 2
N_size = (p*(N-1)+m)+1
k = -(N-1):(N-1)
c = map(complex,zeros(2N_size-1))
ν = 10.0.^(-(1 .+ abs.(k)))
c[p*k .+ m .+ N_size] .= ν
c[p*k .- m .+ N_size] .= ν
d = map(complex,zeros(2N_size-1))
d[p*k .+ m .+ N_size] .= im*ν
d[p*k .- m .+ N_size] .= -im*ν
plot_sol(c,d)
```
